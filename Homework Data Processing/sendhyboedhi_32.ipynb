Dengan menggunakan data `titanic.csv`, buatlah dataframe untuk melihat informasi `max` dan `min` pada column `Age` dan juga informasi `mean` dan `median` pada column `Fare` pada masing-masing PassengerClass atau `Pclass`.

```python
## Your code below here

df_titanic = pd.read_csv('titanic.csv')
df_titanic.head()
 
df_titanic.groupby('Pclass').agg({'Age'  : ['max','min'],
                                  'Fare' : ['mean','median']
                                 })
```

```python
pd.pivot_table(
    df_titanic,
    index=["Pclass"],
    values=["Age", "Fare"],
    aggfunc={"Age": ["max", "min"], "Fare": ["mean", "median"]}
)
```

```python
df_titanic.groupby('Pclass').agg({'Age':['describe']})
```

### 7. Merge DataFrame (25pts)

Buatlah 2 dataframe mengikuti petunjuk berikut:

df_1 :  Filter dataset 01 telecom_revenue.csv dengan multiple kondisi (5pts)
- MonthlyRevenue > 10 
- Occupation terdiri dari Professional, Student, and Crafts

df_2 :  Filter dataset 02 telecom_usage.csv dengan multiple kondisi (5pt)
- nilai UnansweredCalls > BlockedCalls

Jawablah pertanyaan berikut ini:
1. Hitunglah total CustomerID dan rata-rata DroppedCalls untuk masing-masing occupation. (10 Points)
2. Pada occupation apa, rata-rata DroppedCalls paling besar? (5 Points)

**Expected Result:**

![Expected Result](assets/homework_exercise_7_result.png)

```python
## Your code below here
# cek duplikate

df_telecom_revenue = pd.read_csv('01 telecom_revenue.csv')
df_telecom_revenue
```

```python
df_telecom_usage = pd.read_csv('02 telecom_usage.csv')
df_telecom_usage
```

```python
df_1 = df_telecom_revenue[(df_telecom_revenue['MonthlyRevenue'] > 10 ) & (df_telecom_revenue['Occupation'].isin(['Professional','Student','Crafts']))]
df_1
```

```python
df_2 = df_telecom_usage[(df_telecom_usage['UnansweredCalls']) > (df_telecom_usage['BlockedCalls'])]
df_2
```

```python
df_join = df_1.merge(df_2,on='CustomerID', how='inner')
df_join.head(3)
```

```python
df_group_ocp = df_join.groupby('Occupation').agg({
                                    'CustomerID' :['count'],
                                    'DroppedCalls' : ['mean']
                                }).drop_duplicates()
df_group_ocp
```

Jadi rata-rata DroppedCalls terbesar terdapat pada Occupation **Student**
