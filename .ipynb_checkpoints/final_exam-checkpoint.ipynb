```python
import pandas as pd
df_job = pd.read_csv('job_data_analyst_details.csv')
df_job
```

Soal 1 :
Antoni memiki cita-cita menjadi gamer, sehingga dia ingin bekerja di Industry = Video Games yang memiliki Rating perusahaan > 4.5.

Di mana lokasi perusahaan yang memenuhi kriteria Antoni?

```python
df_job[(df_job['Industry'] == 'Video Games')
      & (df_job['Rating'] > 4.5)
      ]
```

Soal 2 :
Pada perusahaan dengan Lokasi yang mengandung kata "Houston",
berapa rata-rata Rating di sana?

```python
df_houston = df_job[(df_job['Location'].str.contains('Houston'))]
df_houston['Rating'].mean()
```

Soal 3 :
Dari perusahaan yang bergerak di Industry = Internet dan Founded > 2015,

Job TItle yang paling banyak lowongan pekerjaanya adalah?

```python
df_job_title = df_job[(df_job['Industry'] == 'Internet')
      & (df_job['Founded'] > 2015)
      ]

df_job_title
```

```python
df_job_title.groupby('Job Title').agg({'Job Title' : 'count'})
```

Soal 4 :
df1 didefinisikan sebagai perusahaan yang Founded = 2018,

df2 didefinisikan sebagai perusahaan yang Founded = 1995.

Kemudian append kedua dataframe tersebut.

```python
df1 = df_job[(df_job['Founded'] == 2018)]

df2 = df_job[(df_job['Founded'] == 1995)]

df_append = pd.concat((df1,df2))
df_append['Job Title'].count()
```

Soal 5 :
Pada perusahaan yang Founded > 2017 , Rating >= 3.5 dan Rating <= 4.5,

Jika dibuat kolom rating_baru yang bernilai Rating+0.5,

maka berapa rata-rata rating_baru tersebut?

```python
df_job5 = df_job[(df_job['Founded'] > 2017)
      & (df_job['Rating'] >= 3.5)
      & (df_job['Rating'] <= 4.5)
      ]
df_job5
```

```python
df_job5['Rating Plus'] = df_job5['Rating'] + 0.5
df_job5
```

```python
df_ratplus = df_job5['Rating Plus'].mean()
df_ratplus
```

Soal 6 :
df1 didefinisikan semua jobs dengan kolom description mengandung kata attitude dan Rating >= 1
df2 didefinisikan semua jobs di New York, NY dan Rating >= 1
append kedua df tersebut, berapa mean dari Rating dari seluruh jobs yang tersedia

```python
df_job
df1_6 = df_job[(df_job['Job Description'].str.contains('attitude'))
               & (df_job['Rating'] >= 1)
              ]

df2_6 = df_job[(df_job['Location'] == 'New York, NY')
              & (df_job['Rating'] >= 1)
              ]

df_append_6 = df1_6.append(df2_6,ignore_index=True)
df_rat_6 = df_append_6['Rating'].mean()
df_rat_6
```
